## El péndulo forzado amorgituado

la ecuación del péndulo forzado amortiguado es

Aceleración+Amortiguamiento+Gravedad=Fuerza externa

Que como ecuación se ve

$$\ddot{\theta}+\mu \dot{\theta}+\sin(\theta)=A \cos (w_d \ t)$$

1.- Límite lineal. Considerando ángulos pequeños, tenemos la ecuación lineal forzada y amortiguada vista en clase, cuya solución particular (aquella que domina a tiempos grandes) tiene una función Coseno con una cierta amplitud y fase dada en términos de los parametros de la ecuación anterior. Tomando $A=1, \omega_0=1, \mu=0.05$

**a) Graficar la Amplitud como función de la frecuencia de la fuerza externa. ¿Exista alguna resonancia?**

<font color=blue>

Tenemos que la solución a la ecuación diferencial es: $$\boxed{\theta(t) = \theta_{hom} + \frac{A}{\sqrt{(\omega_o^2-
\omega^2)^2+\mu^2\omega^2}}\cos(\omega t - \delta)}$$

Donde $$\boxed{\delta = \tan^{-1}\left(\frac{\mu\omega}{\omega_0^2 - \omega^2}\right)} $$


Tomamos solamente la amplitud que está en función de la frecuencia externa $\omega$ y los valores A = 1, $\omega_0 = 1$, $\mu = 0.05 $


</font>


```python
%matplotlib inline

import numpy as np                                       #Importamos numpy como plot
import matplotlib.pyplot as plt                          #Importamos matplotlib como plt

w = np.linspace(0,5,1000)                                 #Valores para la frecuencia externa w, desde 0 hasta 5
mu = 0.05

def A(w):                                                #Definimos la funcion para la amplitud
    return 1/(np.sqrt((1 - w**2)**2 + (w**2)*(mu**2))) 


plt.figure(figsize = (9,6))
plt.title("Amplitud como función de la frecuencia externa", color = "red", fontsize = 20)
plt.ylabel("Amplitud",fontsize = 18)
plt.ylim(0,22.5)
plt.xlabel("Frecuencia", fontsize = 18)
plt.xlim(0,3.5)
plt.grid(True)
plt.plot(w,A(w), color = "orange")
plt.show()

```

<font color=blue>
Para determinar si existe resonancia, buscamos el valor de $\omega$ en el cual la amplitud es máxima , esto es $$\frac{dD}{d\omega} = 0$$
La derivada de la amplitud respecto a la función debe ser igual a cero en algun valor de $\omega$. 

La resonancia se produce cuando la frecuencia externa es igual a la frecuencia natural del sistema. En conclusión si existe resonancia y es en $\omega= 1$.
<br>
Si $\mu < \omega_0$ Existe resonancia.
</font>

**c) Graficar el ángulo y la velocidad angular como función del tiempo para 2 condiciones iniciales distintas.**

<font color=blue>
De la solución homogenea se encuentran tres casos: Infra-amortiguado, amortiguamiento crítico y sobreamortiguado, dado que $\omega_0^2 > \frac{\mu^2}{4}$ nos encontramos en el caso infra-amortiguado. Para este caso la solución viene expresa de la siguiente manera:
<br>    
    
    $$ \boxed{\theta(t) = Ae^{(-\frac{\mu}{2}t)}cos\left(\sqrt{\omega_0^2 - \frac{\mu^2}{4}}t - \epsilon \right)}$$
 

Tomamos de nuevo $\omega_0 = 1$, A = 1, $\mu = 0.05$, Y $\epsilon = 0$ Por simplicidad y no tener un desfase.<br>
<br>
Debemos agregar la solución particular anteriormente vista, de tal manera que nuestra ecuación general queda de la siguiente forma:<br>
<br>
$$ \boxed{\theta(t) = Ae^{(-\frac{\mu}{2}t)}cos\left(\sqrt{\omega_0^2 - \frac{\mu^2}{4}}t - \epsilon \right) + \frac{A}{\sqrt{(\omega_o^2- \omega^2)^2+\mu^2\omega^2)}}\cos\left(\omega t - \tan^{-1}\left(\frac{\mu\omega}{\omega_0^2 - \omega^2}\right)\right)} $$

<br>
Y la velocidad angular:

$$\boxed{\dot{\theta}(t) =-Ae^{-\frac{\mu}{2}t}\left[\frac{\mu}{2}\cdot \cos\left(\sqrt{\omega_0^2 - \frac{\mu^2}{4}}t\right) + \sqrt{\omega_0^2 - \frac{\mu^2}{4}}\cdot\sin\left(\sqrt{\omega_0^2 - \frac{\mu^2}{4}}t\right)\right] + \frac{A\omega}{\sqrt{(\omega_0^2- \omega^2)^2+\mu^2\omega^2)}}\sin\left(\omega t - \delta \right)}  $$

<br>
Tomaremos dos condiciones iniciales la primera $\omega=0.75$ y la segunda $\omega=0.1$
</font>
